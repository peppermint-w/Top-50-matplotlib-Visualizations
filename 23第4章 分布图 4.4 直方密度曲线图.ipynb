横坐标：城市里程<br>

### 4.4.1 导入相应的绘图包

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.sans-serif'] = ['Microsoft Yahei']
plt.rcParams['axes.unicode_minus'] = False
```

### 4.4.2 认识绘制直方密度曲线图的函数

**sns.distplot()**

>重要参数：
>>a：绘制图形的数据<br>
>>bins：直方图参数，把数据分成多少个箱子<br>
>>hist：控制直方图显示的参数(True/False，默认是True)<br>
>>kde：控制密度曲线显示的参数(True/False，默认是True)<br>
>>hist_kws：直方图控制参数<br>
>>kde_kws：密度图控制参数<br>

绘制简单的直方密度曲线图

```python
X = np.random.randn(100)
```

```python
X
```

```python
sns.distplot(X)
```

```python
sns.distplot(X
            ,bins = 10
            #,hist = False #不显示直方图
            #,kde = False #不显示密度曲线
            ,hist_kws = {'color': 'g','histtype':'bar','alpha':0.4} #直方图的属性
            ,kde_kws = {'color':'r','linestyle':'-.','linewidth':3,'alpha':0.7} #密度曲线的属性
            )
```

如果想要把7种类型车的城市里程全部绘制出来，可以使用循环的方式。

```python
df['class'].values
```

```python
np.unique(df['class'].values)
```

```python
clas = np.unique(df['class'].values).tolist()
clas
```

```python
#创建画布
plt.figure(figsize = (14,8), dpi = 60)

#对每一种类型的车进行绘制
for i in clas:
    sns.distplot(df.loc[df['class'] == i, 'cty'].values
                ,label = i
                ,hist = False
                )

#添加装饰
plt.title('城市里程随车类型变化的分布图', fontsize = 20)
plt.ylim(0, 0.65)
plt.xlim(5, 30)
plt.legend(fontsize = 15)
```

绘制compact/suv/minivan这三种车的城市里程分布

```python
#提取出compact/suv/minivan这三种车的城市里程
compact = df.loc[df['class'] == 'compact', 'cty'].values
suv = df.loc[df['class'] == 'suv', 'cty'].values
minivan = df.loc[df['class'] == 'minivan', 'cty'].values

#创建画布
plt.figure(figsize = (14,8), dpi = 60)

#绘制图形
sns.distplot(compact
            ,color = 'b'
            ,label = 'compact'
            ,hist_kws = {'alpha':0.4}
            ,kde_kws = {'linewidth':3, 'alpha':0.7}
            )
sns.distplot(suv
            ,color = 'r'
            ,label = 'suv'
            ,hist_kws = {'alpha':0.4}
            ,kde_kws = {'linewidth':3, 'alpha':0.7}
            )
sns.distplot(minivan
            ,color = 'g'
            ,label = 'minivan'
            ,hist_kws = {'alpha':0.4}
            ,kde_kws = {'linewidth':3, 'alpha':0.7}
            )

#装饰图形
plt.title('compact/suv/minivan三种车的城市里程分布', fontsize = 20)
plt.xlabel('城市里程(cty)', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 15
          ,frameon = False
          )
```

```python
#源代码
#绘制图形
plt.figure(figsize=(13,10), dpi= 80) #创建画布
sns.distplot(df.loc[df['class'] == 'compact', "cty"], color="dodgerblue", label="Compact", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df.loc[df['class'] == 'suv', "cty"], color="orange", label="SUV", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
sns.distplot(df.loc[df['class'] == 'minivan', "cty"], color="g", label="minivan", hist_kws={'alpha':.7}, kde_kws={'linewidth':3})
plt.ylim(0, 0.35) #设定Y轴刻度范围

#装饰图形
plt.title('Density Plot of City Mileage by Vehicle Type', fontsize=22) #添加标题
plt.legend() #添加图例
plt.show() #本地显示图形
```
