### 1.9.1 导入需要的绘图库

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft Yahei']
import seaborn as sns
import warnings
warnings.filterwarnings(action = 'once')
```

```python
large = 22
med = 16
small = 12
params = {
    'axes.titlesize': large,
    'legend.fontsize':med,
    'figure.figsize':med,
    'figure.figsize':(16,10),
    'axes.labelsize':med,
    'axes.titlesize':med,
    'xtick.labelsize':med,
    'ytick.labelsize':med,
    'figure.titlesize':large
}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style('white')
%matplotlib inline
```

### 1.9.2 导入数据

```python
#seaborn中自带数据库，可以从中导入鸢尾花数据集
df = sns.load_dataset('iris')
```

```python
df.columns = ['花萼长', '花萼宽', '花瓣长', '花瓣宽', 'species']
```

### 1.9.3 代码解读，图像解读

**sns.pairplot()**
>重要参数：
>>hue：类别所在的列（和带拟合线的散点图一致），取出数据集的子集，对数据集进行分类<br>
>>kind：绘制的相关性图像类型，可以选择"scatter"散点图或"reg"带拟合线的散点图或"kde"或"hist"<br>
>>diag_kind：对角线图像类型，可以选择"hist"直方图或"kde"<br>
>>corner：True只画下三角

```python
#绘制图像-散点图
plt.figure(figsize = (10,8), dpi = 80) #设置画布
plt.rcParams['font.sans-serif'] = ['Microsoft Yahei']
sns.pairplot(df #数据，各个特征和标签
            ,kind = 'scatter' #要绘制的图像类型
            ,hue = 'species' #类别所在的列
            ,plot_kws = dict(s = 40 #点的尺寸
                            ,edgecolor = 'white' #点的边缘颜色
                            ,linewidth = 1 #点的边缘线宽
                            ) #散点图的一些详细设置
            )
```

```python
#绘制图像-带拟合线
plt.figure(figsize = (10,8), dpi = 80) #设置画布
plt.rcParams['font.sans-serif'] = ['Microsoft Yahei']
sns.pairplot(df #数据，各个特征和标签
            ,kind = 'reg' #要绘制的图像类型
            ,hue = 'species' #类别所在的列
            )
```
