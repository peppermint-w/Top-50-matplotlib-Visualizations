## 1.3 带最佳拟合线的散点图

横坐标：发动机排量(L)<br>纵坐标：公路里程/加仑<br>图例：气缸数量

### 1.3.1 导入需要的绘图库

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 1.3.2 绘制最佳拟合线的函数

<b>sns.lmplot()

```python
X1 = np.linspace(0,10,50)
X1
```

```python
y = 2 * X1 + 5 + np.random.randn(50) * 10 #加上随机数，模拟实现收集数据的时候带来的噪音
```

```python
plt.scatter(X1,y)
```

```python
data = pd.DataFrame({'X1': X1, 'y': y}) #把所有数据放入DataFrame
data
```

```python
gridobj = sns.lmplot('X1', 'y', data = data) #必须通过data+字符串的方式来读取数据
```

可以通过该函数创造一个在数据集的不同子集上拟合回归模型的便捷界面，对于有众多分类特征的回归类数据非常有效。

```python
X2 = [0]*10 + [1]*40
X2
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
data = pd.DataFrame({'X1':X1,'X2':X2,'y':y})
gridobj = sns.lmplot('X1', 'y' #横纵坐标
                     ,data = data #数据
                     ,hue = 'X2' #色调，类似于plt里的c
                     ,legend = False
                    )
plt.legend(['类别0', '类别1'], fontsize = 16)
```

```python
#准备数据
df_select = df.loc[df.cyl.isin([4,8]), :]

#绘制图像
sns.set_style('white') #设立风格
gridobj = sns.lmplot(x = 'displ' #横坐标：发动机排量
                    ,y = 'hwy' #纵坐标：公路里程/加仑
                    ,hue = 'cyl' #分类/子集，气缸数量
                    ,data = df_select #能够输入的数据
                    ,height = 8 #图像的高度
                    ,aspect = 1.6 #图像的纵横比，aspect*height=长度
                    ,palette = 'tab10' #色板tab10(默认)rainbow、summer、tab20等
                    ,legend = False #不显示图例
                    ,scatter_kws = dict(s=60, linewidths=0.7, edgecolors = 'black') #点的尺寸s，点的线宽，点的线的颜色
                    ,truncate = False #不加图像显示不全
                    )

#装饰图像
gridobj.set(xlim=(0.5, 7.5), ylim=(0, 50))
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.xlabel('发动机排量(L)', fontsize = 20)
plt.ylabel('公里路程/加仑', fontsize = 20)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(['气缸数量：4','气缸数量：8'], fontsize = 22)
plt.title('Scatterplot with line of best fit grouped by number of cylinders', fontsize = 20)
plt.show()
```

```python
#一图四线/多图一线col

large = 22
med = 16
small = 12
params = {
    'axes.titlesize': large, #子图上的标题字体大小
    'legend.fontsize': med, #图例字体大小
    'figure.figsize': (16,10), #画布大小
    'axes.labelsize': med, #标签的字体大小
    'xtick.labelsize': med, #x轴上的标尺的字体大小
    'ytick.labelsize': med, #y轴上的标尺的字体大小
    'figure.titlesize': large #整个画布的标题字体大小
}
plt.rcParams.update(params) #设定各种各样的默认属性

#绘制图像
sns.set_style('white') #设立风格
gridobj = sns.lmplot(x = 'displ' #横坐标：发动机排量
                    ,y = 'hwy' #纵坐标：公路里程/加仑
                    ,hue = 'cyl' #分类/子集，气缸数量
                    ,data = df #能够输入的数据
                    ,height = 8 #图像的高度
                    ,robust = True
                    ,aspect = 1.6 #图像的纵横比，aspect*height=长度
                    ,palette = 'tab10' #色板tab10(默认)rainbow、summer、tab20等
                    ,col = 'cyl' #表示按照这个特征中的分类绘制图像，并且一个类别是用一张图一条直线
                    ,col_wrap = 2 #表示对col中的类别数量进行折叠，让一行变多行，每行最多允许3张图
                    #,legend = False #不显示图例
                    ,scatter_kws = dict(s=60, linewidths=0.7, edgecolors = 'black') #点的尺寸s，点的线宽，点的线的颜色
                    ,truncate = False #不加图像显示不全
                    )

#装饰图像
gridobj.set(xlim=(0.5, 7.5), ylim=(0, 50), xlabel = '排量', ylabel = '公路里程/加仑')
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.show()
#图之间加距离还要继续调参
```
