### 4.5.1 导入需要的绘图库

```python
import pandas as pd
import numpy as np
import seaborn as sns
import joypy
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.sans-serif'] = ['Microsoft Yahei']
plt.rcParams['axes.unicode_minus'] = False
```

### 4.5.2 绘制Joyplots的函数

**joypy.joyplot**

函数功能：使用matplotlib和pandas绘制DF或者其他嵌套集合的joyplot

重要参数：

>data：绘图数据集<br>
>column = None：使用data中的有限列进行绘图<br>
>by = None：分组列<br>
>grid = False：添加网格线<br>
>xlabelsize = None：x轴标签大小<br>
>ylabelsize = None：y轴标签大小<br>
>xrot = None：x轴刻度标签旋转角度<br>
>yrot = None：y轴刻度标签旋转角度<br>
>hist = False：显示直方图<br>
>fade = False：如果设定True，则显示渐变色<br>
>ymin = 'max'：共享y轴刻度<br>
>fill = True：曲线下填充颜色<br>
>linecolor = None：曲线颜色<br>
>background = None：背景颜色<br>
>overlap = 1：控制重叠程度<br>
>title = None：添加图表标题<br>
>colormap = None：色谱

### 4.5.3 绘制鸢尾花数据集的joyplots

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris = pd.concat([pd.DataFrame(iris.data),pd.DataFrame(iris.target)], axis = 1)
```

```python
iris.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
```

```python
iris.head()
```

```python
iris.shape
```

```python
joypy.joyplot(iris.iloc[:,0:4])
```

```python
joypy.joyplot(iris
             #,column = ['SepalLength', 'PetalLength'] #选取部分特征
             ,by = 'Species' #分组依据列
             ,xlabelsize = 12 #x轴标尺的字体大小
             ,ylabelsize = 12 #y轴标尺的字体大小
             #,ylim = 'own' #选择各自的刻度，默认是'max'共享y轴刻度
             #,grid = True #添加网格线
             #,fill = False #不填充颜色
             #,hist = True #显示直方图
             #,yrot = 30 #y轴刻度逆旋转30度
             #,xrot = 60 #x轴刻度逆旋转36度
             #,fade = True #渐变色，从上到下依次变身
             #,overlap = 3 #数值越小，y轴间隔越大
             #,linecolor = 'b' #线的颜色
             #,background = 'grey' #背景颜色
             ,title = 'Joyplots of iris' #标题
             ,colormap = plt.cm.rainbow #填充颜色
             )
```

### 4.5.5 绘制Joyplots

```python
df['class'].value_counts()
```

```python
#创建画布
plt.figure(figsize=(16,10), dpi = 60)

#绘制图像
fig, axes = joypy.joyplot(df
                         ,column = ['hwy', 'cty'] #蓝色是hwy
                         ,by = 'class'
                         ,ylim = 'own'
                         ,figsize = (10,6)
                         ,xlabelsize = 12
                         ,ylabelsize = 12
                         ,title = 'Joy Plot of City and Highway Mileage by Class'
                         )
```

```python
#源码
#Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
fig, axes = joypy.joyplot(df, column=['hwy', 'cty'], by="class", ylim='own', figsize=(14,10))

#Decoration
plt.title('Joy Plot of City and Highway Mileage by Class', fontsize=22)
plt.show()
```

### 4.5.6 图像解读

对于同一种车型，公路里程比城市里程大，即消耗一加仑的油，公路比城市跑得多。<br>
对于不同类型的车，皮卡的城市里程和公路里程较小，即消耗一加仑的有，皮卡在公路和城市跑得少。
